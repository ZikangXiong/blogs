# Optimal Vehicle Path Planning Using Quadratic Optimization for Baidu Apollo Open Platform
[link](https://ieeexplore.ieee.org/document/9304787)  

## Summary


The optimization only focuses on the lateral direction. 

1. Generate a smooth guided line. This line acts as the bridge frenet Frame and map. (Why not use the lane on road?)
2. Changing lanes decision is based on a rule-based decision tree. 
3. With the decisions from 2 and the location information of obstacles generate a set of feasible regions (geometrical homotopy) for optimization. 
4. Optimize with Piecewise-jerk, with the safety constraints from 3 and kinematic constraints.

## Hyper-parameters

Minimize the following function with safety and kinematic constraints in paper.  
$\begin{aligned} \tilde{\boldsymbol{f}}(\boldsymbol{l}(s)) &=w_{l} * \sum_{i=0}^{n-1} l_{i}^{2}+w_{l^{\prime}} * \sum_{i=0}^{n-1} l_{i}^{\prime 2}+w_{l^{\prime \prime}} * \sum_{i=0}^{n-1} l_{i}^{\prime \prime 2} \\ &+w_{l^{\prime \prime \prime}} * \sum_{i=0}^{n-2}\left(\frac{l_{i+1}^{\prime \prime}-l_{i}^{\prime \prime}}{\Delta s}\right)^{2} \\ &+w_{o b s} * \sum_{i=0}^{n-1}\left(l_{i}-0.5 *\left(l_{\min }^{i}+l_{\max }^{i}\right)\right)^{2} \end{aligned}$

- $w_l$: Weight of lateral deviation. 
- $w_l'$: Weight of lateral movement. 
- $w_l''$: Weight of change of lateral movement.
- $w_l'''$: Weight of jerk.
- $w_{obs}$: Weight of distance to unsafe region.
