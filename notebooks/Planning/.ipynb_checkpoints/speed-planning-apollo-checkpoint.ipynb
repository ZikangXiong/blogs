# Optimal Trajectory Generation for Autonomous Vehicles Under Centripetal Acceleration Constraints for In-lane Driving Scenarios
[link](https://ieeexplore.ieee.org/document/8916917)   

## Summary
  
This works only considers the longitudinal optimization in single lane. 

1. Generate a guideline using the polynomial spiral curve as piecewise basis, and smooth it with quintic polynomial spiral curves connection. The guideline is the bridge between Cartesian coordinate and Frenet Frame.  
2. Optimize the trajectory with the above goals with constraints generated with path-time-obstacle and constraints on longitudinal velocity, longitudinal acceleration, centripetal acceleration, and longitudinal jerk.  

## Hyper-parameters
Minimize the following equation with constraints in the paper. 



These weights corresponse to  
$w_{\ddot{s}}$: longitudinal acceleration     
$w_{\dddot{s}}$: longitudinal jerk  
$w_{a_c}$: centripetal acceleration  
$w_{\dot{s}_{r e f}} $: Distance to arrive referance speed       
$w_{s_{\text{task}}} $: Distance between final location to the predefeind state.    
$w_{\dot{s}_{\text {task}}}$: Distance between final velocity to the predefeind state.  
$w_{\ddot{s}_{\text {task}}}$: Distance between final acceleration to the predefeind state.
