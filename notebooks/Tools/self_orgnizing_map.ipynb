# Self-orgnizing Map
[blog](https://www.kaggle.com/residentmario/self-organizing-maps)  
[code](https://github.com/JustGlowing/minisom)  

The intuition behaind is 
1. The neural network outputs vectors in lower dimension.
2. Read the input one-by-one, and compute there representation in lower dimension. 
3. Find the closest vector and move the representation towards the closest vector with gradient.

```python
data = [[ 0.80,  0.55,  0.22,  0.03],
        [ 0.82,  0.50,  0.23,  0.03],
        [ 0.80,  0.54,  0.22,  0.03],
        [ 0.80,  0.53,  0.26,  0.03],
        [ 0.79,  0.56,  0.22,  0.03],
        [ 0.75,  0.60,  0.25,  0.03],
        [ 0.77,  0.59,  0.22,  0.03]]  
```

```python
from minisom import MiniSom    
som = MiniSom(6, 6, 4, sigma=0.3, learning_rate=0.5) # initialization of 6x6 SOM
som.train(data, 100) # trains the SOM with 100 iterations
```

```python
som.winner(data[-1])
```
